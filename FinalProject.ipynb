```python
# ── Format exactly as the paper did ───────────────────────────
def format_recipe(recipe):
    """
    Format a recipe dict into a single string the way the paper did:
    'Title: X, Ingredients: a, b, c, Steps: Step 1: ... Step 2: ...'
    """
    title = recipe.get("title", "")
    ingredients = ", ".join(recipe.get("ingredients", []))
    steps = " ".join(recipe.get("steps", []))
    return f"Title: {title}, Ingredients: {ingredients}, Steps: {steps}"

# Show formatted versions of source and first target
record = datasets["recipe_match_cn2en"]["train"][0]
print("=== FORMATTED SOURCE (Chinese) ===")
print(format_recipe(record["source"]))
print("\n=== FORMATTED TARGET (English) ===")
print(format_recipe(record["targets"][0]))
```

```python
# ── Flatten to 1:1 pairs as the paper did ─────────────────────
def flatten_dataset(records):
    """Convert 1:many matched recipes into 1:1 source-target pairs."""
    pairs = []
    for record in records:
        source_str = format_recipe(record["source"])
        for target in record["targets"]:
            target_str = format_recipe(target)
            pairs.append({
                "source": source_str,
                "target": target_str
            })
    return pairs

# Flatten all splits for both directions
flat = {}
for direction in ["recipe_match_cn2en", "recipe_match_en2cn"]:
    flat[direction] = {}
    for split in ["train", "valid", "test"]:
        flat[direction][split] = flatten_dataset(datasets[direction][split])
        print(f"{direction}/{split}: {len(flat[direction][split])} 1:1 pairs")
```

```python
# ── Create HuggingFace Dataset objects ─────────────────
from datasets import Dataset, DatasetDict
import json, random

def load_flat_jsonl(path):
    records = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                records.append(json.loads(line))
    return records

save_dir = "/content/drive/MyDrive/DS266/cultural-recipes-data/flattened"

# ── Subset strategy for Colab ──────────────────────────────────
# Stage 1 (test run):   USE_SUBSET=True,  SUBSET_SIZE=5000
# Stage 2 (full run):   USE_SUBSET=False
USE_SUBSET  = True
SUBSET_SIZE = 96149

random.seed(42)
raw_datasets = {}

for direction in ["recipe_match_cn2en", "recipe_match_en2cn"]:
    train = load_flat_jsonl(f"{save_dir}/{direction}/train.jsonl")
    valid = load_flat_jsonl(f"{save_dir}/{direction}/valid.jsonl")
    test  = load_flat_jsonl(f"{save_dir}/{direction}/test.jsonl")

    if USE_SUBSET:
        train = random.sample(train, min(SUBSET_SIZE, len(train)))
        valid = random.sample(valid, min(12469,        len(valid)))
        test  = random.sample(test,  min(12056,        len(test)))

    raw_datasets[direction] = DatasetDict({
        "train": Dataset.from_list(train),
        "valid": Dataset.from_list(valid),
        "test":  Dataset.from_list(test),
    })
    print(f"{direction}: train={len(train):,}, "
          f"valid={len(valid):,}, test={len(test):,}")
```

```python
MAX_INPUT_LENGTH  = 512
MAX_TARGET_LENGTH = 512

def tokenize(batch):
    model_inputs = tokenizer(
        batch["source"],
        max_length=MAX_INPUT_LENGTH,
        truncation=True,
        # NO padding here — data collator handles it dynamically
    )
    labels = tokenizer(
        text_target=batch["target"],
        max_length=MAX_TARGET_LENGTH,
        truncation=True,
        # NO padding here either
    )
    # Replace padding token id with -100 so loss ignores it
    labels["input_ids"] = [
        [(l if l != tokenizer.pad_token_id else -100) for l in label]
        for label in labels["input_ids"]
    ]
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

tokenized = {}
for direction in ["recipe_match_cn2en", "recipe_match_en2cn"]:
    tokenized[direction] = raw_datasets[direction].map(
        tokenize,
        batched=True,
        batch_size=64,
        remove_columns=["source", "target"],
    )
    print(f"Tokenized {direction}: {tokenized[direction]}")
```

```python
# ── Step 6: Combine both directions ───────────────────────────
from datasets import concatenate_datasets

combined = DatasetDict({
    "train": concatenate_datasets([
        tokenized["recipe_match_cn2en"]["train"],
        tokenized["recipe_match_en2cn"]["train"],
    ]),
    "valid": concatenate_datasets([
        tokenized["recipe_match_cn2en"]["valid"],
        tokenized["recipe_match_en2cn"]["valid"],
    ]),
    "test": concatenate_datasets([
        tokenized["recipe_match_cn2en"]["test"],
        tokenized["recipe_match_en2cn"]["test"],
    ]),
})

print(f"\nCombined dataset:")
print(f"  train: {len(combined['train']):,} pairs")
print(f"  valid: {len(combined['valid']):,} pairs")
print(f"  test:  {len(combined['test']):,} pairs")
```

```python
#Evaluation
# ── Cell 11: Evaluate on test set ─────────────────────────────
import json, torch
from torch.utils.data import DataLoader
from transformers import DataCollatorForSeq2Seq
import sacrebleu
from rouge_score import rouge_scorer

model.eval()

def load_flat_jsonl(path):
    records = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                records.append(json.loads(line))
    return records

save_dir = "/content/drive/MyDrive/DS266/cultural-recipes-data/flattened"

# ── Evaluate on both directions ────────────────────────────────    #Make updates here too
results = {}

for direction in ["recipe_match_cn2en", "recipe_match_en2cn"]:
    print(f"\n=== Evaluating {direction} ===")

    # Load a sample of test pairs (full test set takes very long)
    test_pairs = load_flat_jsonl(f"{save_dir}/{direction}/test.jsonl")

    # Sample 200 for speed — increase for more accurate estimate
    import random
    random.seed(42)
    sample = random.sample(test_pairs, min(500, len(test_pairs)))

    sources    = [r["source"] for r in sample]
    references = [r["target"] for r in sample]

    # ── Generate predictions ───────────────────────────────────
    predictions = []
    batch_size = 4

    for i in range(0, len(sources), batch_size):
        batch_sources = sources[i:i+batch_size]
        inputs = tokenizer(
            batch_sources,
            return_tensors="pt",
            max_length=512,
            truncation=True,
            padding=True,
        ).to(model.device)

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                #max_length=256,
                max_new_tokens=256,
                num_beams=4,
                early_stopping=True,
                no_repeat_ngram_size=4,
                repetition_penalty=2.5,
            )

        decoded = tokenizer.batch_decode(
            outputs, skip_special_tokens=True
        )
        predictions.extend(decoded)

        if (i // batch_size) % 10 == 0:
            print(f"  Generated {min(i+batch_size, len(sources))}"
                  f"/{len(sources)} predictions...")

    # ── Compute BLEU ───────────────────────────────────────────
    # sacrebleu expects list of predictions and list of reference lists
    bleu = sacrebleu.corpus_bleu(
        predictions,
        [references],
        tokenize="char" if direction == "recipe_match_en2cn" else "13a"
    )

    # ── Compute ROUGE ──────────────────────────────────────────
    scorer = rouge_scorer.RougeScorer(
        ["rouge1", "rouge2", "rougeL"], use_stemmer=False
    )

    rouge1_scores, rouge2_scores, rougeL_scores = [], [], []
    for pred, ref in zip(predictions, references):
        scores = scorer.score(ref, pred)
        rouge1_scores.append(scores["rouge1"].fmeasure)
        rouge2_scores.append(scores["rouge2"].fmeasure)
        rougeL_scores.append(scores["rougeL"].fmeasure)

    results[direction] = {
        "bleu":   round(bleu.score, 2),
        "rouge1": round(sum(rouge1_scores) / len(rouge1_scores) * 100, 2),
        "rouge2": round(sum(rouge2_scores) / len(rouge2_scores) * 100, 2),
        "rougeL": round(sum(rougeL_scores) / len(rougeL_scores) * 100, 2),
    }

    print(f"\n  Results for {direction}:")
    print(f"    BLEU:   {results[direction]['bleu']}")
    print(f"    ROUGE-1: {results[direction]['rouge1']}")
    print(f"    ROUGE-2: {results[direction]['rouge2']}")
    print(f"    ROUGE-L: {results[direction]['rougeL']}")
```
